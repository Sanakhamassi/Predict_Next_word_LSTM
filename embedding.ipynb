```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[ 'the glass of milk',
'the glass of juice',
'the cup of tea',
'I am a good boy',
'I am a good developer',
'understand the meaning of words',
'your videos are good' ] 
```

```python
# define the vocabulary 
voc_size=10000
```

```python
#on hot represnetation for every worf
one_hot_rep=[one_hot(words,voc_size) for words in sent]
```

```python
sent
one_hot_rep
```

```python
#word embedding representation 
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences 
from tensorflow.keras.models import Sequential
```

```python
import numpy as np 
```

```python
sent_len=8 
embedded_doc=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_len) 
print(embedded_doc)
```

```python
#feature representation 
dim=10 
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len)) 
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
